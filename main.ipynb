```python
import numpy as np;
import matplotlib.pyplot as plt;


def gen_data(n, ep_mean, ep_var, x_mean, x_var):
  np.random.seed(90)
  x = np.random.normal(size = n, loc = x_mean, scale = x_var)
  a = 3
  b = 6
  epsilon = np.random.normal(size = n, loc = ep_mean, scale = ep_var)
  y = a * x + b + epsilon
  return np.vstack([np.ones(x.shape), x]).T, y

x_1 , y_1 = gen_data(100, 1, 0.2, 1, 0.3)
x_2 , y_2 = gen_data(100, 1, 0.2, 1, 1)
x_3 , y_3 = gen_data(100, 1, 1, 1, 0.3)
x_4 , y_4 = gen_data(100, 1, 1, 1, 1)

fig, axs = plt.subplots(2, 2)
axs[0, 0].scatter(x_1[:, 1], y_1, s = 2)
axs[0, 0].set_title('Low var')
axs[0, 1].scatter(x_2[:, 1], y_2, s = 2)
axs[0, 1].set_title('High x var')
axs[1, 0].scatter(x_3[:, 1], y_3, s = 2)
axs[1, 0].set_title('High y var')
axs[1, 1].scatter(x_4[:, 1], y_4, s = 2)
axs[1, 1].set_title('Both high var')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

for ax in axs.flat:
    ax.label_outer()
```

```python
def get_otlinere_color(x):
  U, S, Vh = np.linalg.svd(x, full_matrices=False)
  H = np.matmul(U, U.T).diagonal()
  threshold = np.percentile(H, 90)
  color = np.where(H > threshold, 'r', 'g')
  return color

color_1 = get_otlinere_color(x_1)
color_2 = get_otlinere_color(x_2)
color_3 = get_otlinere_color(x_3)
color_4 = get_otlinere_color(x_4)

fig, axs = plt.subplots(2, 2)
axs[0, 0].scatter(x_1[:, 1], y_1, s = 2, color = color_1)
axs[0, 0].set_title('Low var')
axs[0, 1].scatter(x_2[:, 1], y_2, s = 2, color = color_2)
axs[0, 1].set_title('High x var')
axs[1, 0].scatter(x_3[:, 1], y_3, s = 2, color = color_3)
axs[1, 0].set_title('High y var')
axs[1, 1].scatter(x_4[:, 1], y_4, s = 2, color = color_4)
axs[1, 1].set_title('Both high var')
```

```python
import numpy as np;
import matplotlib.pyplot as plt;


def gen_data_3D(n, ep_mean, ep_var, x_mean, x_var):
  np.random.seed(90)
  x_1 = np.random.normal(size = n, loc = x_mean, scale = x_var)
  x_2 = np.random.normal(size = n, loc = x_mean, scale = x_var)
  a = 3
  b = 6
  c = 10
  epsilon = np.random.normal(size = n, loc = ep_mean, scale = ep_var)
  y = a * x_1 + b * x_2 + c + epsilon
  return np.vstack([np.ones(x_1.shape), x_1, x_2]).T, y

x_1 , y_1 = gen_data_3D(100, 1, 0.2, 1, 0.3)
x_2 , y_2 = gen_data_3D(100, 1, 0.2, 1, 1)
x_3 , y_3 = gen_data_3D(100, 1, 1, 1, 0.3)
x_4 , y_4 = gen_data_3D(100, 1, 1, 1, 1)

fig, axs = plt.subplots(2, 2, subplot_kw={"projection": "3d"}, figsize=(10, 5))
axs[0, 0].scatter3D(x_1[:, 1:2], y_1, s = 4)
axs[0, 0].set_title('Low var')
axs[0, 1].scatter3D(x_2[:, 1:2], y_2, s = 4)
axs[0, 1].set_title('High x var')
axs[1, 0].scatter3D(x_3[:, 1:2], y_3, s = 4)
axs[1, 0].set_title('High y var')
axs[1, 1].scatter3D(x_4[:, 1:2], y_4, s = 4)
axs[1, 1].set_title('Both high var')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

for ax in axs.flat:
    ax.label_outer()
```

```python
import pyod as py
from pyod.models.knn import KNN
from sklearn.metrics import balanced_accuracy_score

def colors(arr):
    return np.where(arr == 0, 'g', 'r')

x_train, x_test, y_train, y_test = py.utils.data.generate_data_clusters(
    n_train=400,
    n_test=200,
    n_features=2,
    contamination=0.1,
    random_state=33,
    n_clusters=2
)

knn = KNN()
knn.fit(x_train)
y_test_pred = knn.predict(x_test)

ba = balanced_accuracy_score(y_test, y_test_pred)
print("Balanced Accuracy:", ba)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].scatter(x_train[:,0], x_train[:,1], c = colors(y_train))
axs[0, 0].set_title("Ground truth training")

axs[0, 1].scatter(x_train[:,0], x_train[:,1], c = colors(knn.labels_))
axs[0, 1].set_title("Predicted training")

axs[1, 0].scatter(x_test[:,0], x_test[:,1], c = colors(y_test))
axs[1, 0].set_title("Ground truth test")

axs[1, 1].scatter(x_test[:,0], x_test[:,1], c = colors(y_test_pred))
axs[1, 1].set_title("Predicted test")

plt.tight_layout()
plt.show()
```

```python
import scipy.io as sio
from sklearn.model_selection import train_test_split

odds = sio.loadmat('cardio.mat')
x = odds['X']
x_min = np.min(x, axis=0)
x_max = np.max(x, axis=0)
x = (x - x_min) / (x_max - x_min) #normalization
y = odds['y']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=33)
```

```python
lof_models = []
ba_scores = []
for i in range(10):
    n_neighbors = 30 + i * 20
    lof_model = LOF(n_neighbors = n_neighbors)
    lof_model.fit(x_train)
    lof_models.append(lof_model)
    ba_train = balanced_accuracy_score(y_train, lof_model.labels_)
    ba_test = balanced_accuracy_score(y_test, lof_model.predict(x_test))
    print('BA train', i, ':', ba_train)
    print('BA test', i, ':', ba_test)
    ba_scores.append([ba_train, ba_test])
```

```python
from pyod.utils.utility import standardizer
from pyod.models.combination import average
from pyod.models.combination import maximization

ba_scores_standardized = standardizer(ba_scores)
ba_scores_average = average(ba_scores_standardized)
ba_scores_maximization = maximization(ba_scores_standardized)
print("BA average:", ba_scores_average)
print("BA maximization:", ba_scores_maximization)
```

```python
average_mean = np.mean(ba_scores_average)
average_deviation = np.std(ba_scores_average)
average_z_score = np.max(abs(ba_scores_average - average_mean) / average_deviation)
average_quantile = np.quantile(average_z_score, 0.9)
print("Average strategy quantile:", average_quantile)

maximization_mean = np.mean(ba_scores_maximization)
maximization_deviation = np.std(ba_scores_maximization)
maximization_z_score = np.max(abs(ba_scores_maximization -maximization_mean) / maximization_deviation)
maximization_quantile = np.quantile(maximization_z_score, 0.9)
print("Maximization strategy quantile", maximization_quantile)
```
